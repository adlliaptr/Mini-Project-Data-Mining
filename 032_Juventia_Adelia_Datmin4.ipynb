####***Penanganan Outliers***

#####Pengecekkan Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan tabel diatas, dapat dilihat bahwa ketiga kolom tersebut memiliki nilai outliers kurang dari 10%. Hal ini mengidentifikasikan bahwa tidak ada data yang memiliki jangkauan yang terlalu jauh dari data-data lainnya.

#####Imputasi Outliers

Melakukan flooring dan capping pada kolom 'math score', 'reading score', dan 'writing score'

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
df['math score']=np.where(df['math score']>upper_whisker,upper_whisker,np.where(df['math score']<lower_whisker,lower_whisker,df['math score']))
```

```python
Q1 = df['reading score'].quantile(0.25)
Q3 = df['reading score'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
df['reading score']=np.where(df['reading score']>upper_whisker,upper_whisker,np.where(df['reading score']<lower_whisker,lower_whisker,df['reading score']))
```

```python
Q1 = df['writing score'].quantile(0.25)
Q3 = df['writing score'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
df['writing score']=np.where(df['writing score']>upper_whisker,upper_whisker,np.where(df['writing score']<lower_whisker,lower_whisker,df['writing score']))
```

Syntax diatas digunakan untuk mengatasi outliers dalam data. Syntax tersebut menggunakan quartil pertama (Q1) dan quartil ketiga (Q3) untuk menghitung rentang antarquartil (IQR). Nilai lower whisker dan upper whisker akan dihitung menggunakan IQR dan faktor pemanjangan 1.5.

```python
df
```

```python
def score_category_math(score):
    if score < 25:
        return 'Beginner'
    elif 25 <= score <= 45:
        return 'Elementary'
    elif 45 < score <= 75:
        return 'Intermediate'
    elif 75 < score <= 90:
        return 'Advanced'
    else:
        return 'Master'

df['score_category_math'] = df['math score'].apply(score_category_math)
```

```python
def score_category_reading(score):
    if score < 25:
        return 'Beginner'
    elif 25 <= score <= 45:
        return 'Elementary'
    elif 45 < score <= 75:
        return 'Intermediate'
    elif 75 < score <= 90:
        return 'Advanced'
    else:
        return 'Master'

df['score_category_reading'] = df['reading score'].apply(score_category_reading)
```

```python
def score_category_writing(score):
    if score < 25:
        return 'Beginner'
    elif 25 <= score <= 45:
        return 'Elementary'
    elif 45 < score <= 75:
        return 'Intermediate'
    elif 75 < score <= 90:
        return 'Advanced'
    else:
        return 'Master'

df['score_category_writing'] = df['writing score'].apply(score_category_writing)
```

Syntax diatas digunakan untuk mengkategorikan nilai matematika, nilai menulis, dan nilai membaca kedalam berbagai kategori sesuai dengan rendang nilainya. Setiap kategori memiliki rentang nilai yang spesifik. Contohnya, kategori Beginner untuk nlai kurang dari 25. Jategori Elementary untuk nilai darikurang atau lebih dari 25 hingga 45. Dan seterusnya. Hasil kategori ini akan di masukkan ke dalam kolom baru yang disebut **score_category_reading, score_category_writing, dan score_category_math**

#**Modelling Predict**

##**Pre Processing**

###**Split Data**

```python
x = df.drop('math score', axis=1)
y = df['math score']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

Splitting Data adalah proses membagi dataset menjadi subset yang berbeda untuk tujuan tertentu. disini saya menggunakan kolom math score untuk menjadi kolom target.

```python
x_test
```

```python
y_test
```

###**Normalisasi Data**

```python
scaler = MinMaxScaler()

x_train_norm = scaler.fit_transform(x_train)

x_test_norm = scaler.transform(x_test)
```

pada syntax diatas berfungsi untuk melakukan normalisasi data menggunakan metode Min-Max Scalling yaitu mengubah nilai-nilai dalam kolom menjadi rentang tertentu, antara 0 dan 1

```python
x_train_norm
```

```python
x_test_norm
```

##**Build and Train Model**

####**Gaussian Naive Bayes**

```python
gnb = GaussianNB()
gnb.fit(x_train_norm, y_train)
```

Gaussian Naive Bayes adalah salah satu jenis algoritma klasifikasi yang berdasarkan teorema bayes dengan asumsi bahwa nilai fitur dalam setiap kelas mengikuti distribusi Gaussian (normal).

####**K - Nearest Neighbor**

```python
knn = KNeighborsClassifier()

knn.fit(x_train_norm, y_train)
```

Algoritma K-Nearest Neighbors (KNN) adalah salah satu algoritma yang paling sederhana dan populer dalam machine learning, terutama dalam konteks klasifikasi dan regresi. Prinsip dasarnya adalah melakukan prediksi berdasarkan data paling serupa (terdekat) dalam ruang fitur.

####**Decision Tree**

```python
dtc = DecisionTreeClassifier()

dtc.fit(x_train_norm, y_train)
```


Decision Tree (Pohon Keputusan) adalah model prediktif dalam machine learning yang mengambil bentuk struktur pohon. Model ini dapat digunakan baik untuk masalah klasifikasi maupun regresi.

##**Begins and Compare Predict**

```python
gnb_pred = gnb.predict(x_test_norm)
knn_pred = knn.predict(x_test_norm)
dtc_pred = dtc.predict(x_test_norm)
```

gnb_pred = gnb.predict(x_test_norm): Pada baris ini, dilakukan prediksi menggunakan model Gaussian Naive Bayes (gnb) yang telah dilatih sebelumnya. Fungsi predict() digunakan untuk melakukan prediksi terhadap data pengujian yang sudah dinormalisasi (x_test_norm). Hasil prediksi akan disimpan dalam variabel gnb_pred.

knn_pred = knn.predict(x_test_norm): Pada baris ini, dilakukan prediksi menggunakan model K-Nearest Neighbors (knn) yang telah dilatih sebelumnya. Fungsi predict() juga digunakan untuk melakukan prediksi terhadap data pengujian yang sudah dinormalisasi (x_test_norm). Hasil prediksi akan disimpan dalam variabel knn_pred.

dtc_pred = dtc.predict(x_test_norm): Pada baris ini, dilakukan prediksi menggunakan model Decision Tree Classifier (dtc) yang telah dilatih sebelumnya. Fungsi predict() digunakan untuk melakukan prediksi terhadap data pengujian yang sudah dinormalisasi (x_test_norm). Hasil prediksi akan disimpan dalam variabel dtc_pred.

```python
x_test = pd.DataFrame(x_test).reset_index(drop=True)

y_test = pd.DataFrame(y_test).reset_index(drop=True)

gnb_col = pd.DataFrame(gnb_pred.astype(int), columns=["gnb_prediction"])
knn_col = pd.DataFrame(knn_pred.astype(int), columns=["knn_prediction"])
dtc_col = pd.DataFrame(dtc_pred.astype(int), columns=["dtc_prediction"])

combined_data = pd.concat([x_test, y_test, gnb_col, knn_col, dtc_col], axis=1)
```

x_test = pd.DataFrame(x_test).reset_index(drop=True): Baris ini membuat DataFrame Pandas baru dari data pengujian (x_test) dan kemudian mengatur ulang indeksnya agar dimulai dari 0. Ini berguna untuk memastikan bahwa indeks dari DataFrame x_test dimulai dari 0 dan berurutan.

y_test = pd.DataFrame(y_test).reset_index(drop=True): Baris ini melakukan hal yang sama dengan baris sebelumnya, tetapi untuk data label (y_test). Ini juga mengubah data label menjadi DataFrame Pandas dan mengatur ulang indeksnya agar dimulai dari 0.

gnb_col = pd.DataFrame(gnb_pred.astype(int), columns=["gnb_prediction"]): Baris ini membuat DataFrame baru dari prediksi yang dihasilkan oleh model Gaussian Naive Bayes (gnb_pred). Fungsi astype(int) digunakan untuk mengubah tipe data prediksi menjadi integer, dan kemudian nama kolomnya diatur sebagai "gnb_prediction".

knn_col = pd.DataFrame(knn_pred.astype(int), columns=["knn_prediction"]): Baris ini membuat DataFrame baru dari prediksi yang dihasilkan oleh model K-Nearest Neighbors (knn_pred). Seperti sebelumnya, tipe data prediksi diubah menjadi integer dan nama kolomnya diatur sebagai "knn_prediction".

dtc_col = pd.DataFrame(dtc_pred.astype(int), columns=["dtc_prediction"]): Baris ini membuat DataFrame baru dari prediksi yang dihasilkan oleh model Decision Tree Classifier (dtc_pred). Prediksi juga diubah menjadi tipe data integer dan nama kolomnya diatur sebagai "dtc_prediction".

combined_data = pd.concat([x_test, y_test, gnb_col, knn_col, dtc_col], axis=1): Baris ini menggabungkan semua DataFrame yang telah dibuat sebelumnya (x_test, y_test, gnb_col, knn_col, dtc_col) menjadi satu DataFrame tunggal yang disebut combined_data. Penggabungan dilakukan secara horizontal (axis=1), sehingga kolom dari masing-masing DataFrame digabungkan. Ini memungkinkan kita untuk melihat fitur-fitur, label aktual, dan prediksi dari tiga model yang berbeda dalam satu DataFrame untuk analisis lebih lanjut.
